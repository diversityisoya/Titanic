```python
#ここから課題の処理
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna("S")

#欠損した年齢データを中央値で補完(平均値の方がよいという話もある)
#欠損した出港地をSで補完
```

```python
train["Sex"][train["Sex"] == "male"]=0
train["Sex"][train["Sex"] == "female"]=1
train["Embarked"][train["Embarked"] == "S"]=0
train["Embarked"][train["Embarked"] == "C"]=1
train["Embarked"][train["Embarked"] == "Q"]=2
#カテゴリカルデータの文字列を数値に変換
#警告出てるけど実行はできている

train.head(10)
```

```python
train.isnull().sum()
```

```python
from sklearn import tree
#scikit-learnをインポート
```

```python
target = train["Survived"].values
#trainの正解ラベルの取得
```

```python
features_two = train[["Pclass","Age","Sex","Fare","SibSp","Parch","Embarked"]].values

#features_two = train.drop(["Survived","PassengerId","Name","Ticket"],1)
#でもOK,第一引数がラベル,第二引数は0->行,1->列を指定

#追加となった項目も含めて予測モデルその2で使う値を取り出す
#使用カラムはPclass,Age,Sex,Fare,SibSp,Parch,Embarked
#さらに工夫するのであれば、
#SibSp+Parchで
#家族と乗船したかどうかが分かって、生存者推定に使いやすいデータになる。
#Cabin(客室番号)は先頭アルファベットを参照して数値に変換することができる。
#根本的に手法を変えるなら、SVMやランダムフォレストを使うとさらに精度が良くなる。

max_depth = 10
min_samples_split = 5
my_tree_two = tree.DecisionTreeClassifier(max_depth = max_depth,min_samples_split=min_samples_split,random_state = 1)
#my_tree_two = my_tree_two.fit(features_two,target)
#決定木の作成とアーギュメントの設定
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(my_tree_two,features_two,target,cv=5)
print(scores)
#交差検証
#今回は単純な決定木を使用しているので実行結果は一意に定まる
#ランダムフォレストではそうはならないらしい
```
